### *Request name and bill for each person*

```julia
function splitBill(number_of_friends)
    name = []
    bill = []
    subtotal = 0
    
    for i in 1:number_of_friends
        if i != 1
            print("What is your friend's name?\n")
            flush(stdout)
        else
            print("What is your name?\n")
            flush(stdout)
        end
        push!(name,readline())
        print(name[i],"\n\n")
        if i != 1
            print("How much was ",name[i],"'s bill?\n")
            flush(stdout)
        else
            print("How much was your bill?\n")
            flush(stdout)
        end
        push!(bill,parse(Float32, readline()))
        # Calculates the total cost before the tip is added
        subtotal += bill[i]
        @printf("\$%.2f",bill[i])
        print("\n")
        println("________________________________________________")
        print("\n")
            
    end

    # Asks the user what tip percentage they will pay
    print("What percentage of the total bill would you like to tip?\n")
    flush(stdout)
    tip_perc = parse(Int32,readline())
    println(tip_perc,"%\n")

    @printf("The cost before tipping is: \$%.2f",subtotal)
    print("\n")

    # Calculates the total cost of the food plus tip
    total = round(subtotal*(1+(tip_perc/100)),digits=2);
    @printf("\nThe total cost after tipping is: \$%.2f",total)
    print("\n")    

    # Determines if the bill would be split unevenly
    extra_person = round(total%0.05,digits=2)
    # If the bill is uneven then the user is pays for the extra 
    # cost ($0.04 at most)
    # Else, the bill is split evenly between the user and their # 4 friends
    if extra_person != 0.05
        per_person = (total-extra_person)/5
        @printf("\nThe total cost for my friends is: \$%.2f",per_person)
        @printf("\nThe total for me is: \$%.2f",per_person+extra_person)
    else
        per_person = total/5
        @printf("\nThe total cost per person is: \$%.2f",per_person)
    end

end
```
