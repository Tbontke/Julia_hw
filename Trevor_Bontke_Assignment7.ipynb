```julia
# Allows the use of @printf method for float formating
using Printf

function splitBill(number_of_friends)
    record = Dict()
    record["subtotal"] = 0
    
    for i in 1:number_of_friends
        if i != 1
            print("What is your friend's name?\n")
            flush(stdout)
        else
            print("What is your name?\n")
            flush(stdout)
        end
        record["name"]=readline()
        print(record["name"],"\n\n")
        if i != 1
            print("How much was ",record["name"],"'s bill?\n")
            flush(stdout)
        else
            print("How much was your bill?\n")
            flush(stdout)
        end
        record["bill"]=parse(Float32, readline())
        record[record["name"]]=record["bill"]
        # Calculates the total cost before the tip is added
        record["subtotal"] += record["bill"]
        @printf("\$%.2f",record["bill"])
        print("\n")
        println("________________________________________________")
        print("\n")
            
    end

    # Asks the user what tip percentage they will pay
    print("What percentage of the total bill would you like to tip?\n")
    flush(stdout)
    record["tip_perc"] = parse(Int32,readline())
    println(record["tip_perc"],"%\n")

    @printf("The cost before tipping is: \$%.2f",record["subtotal"])
    print("\n")

    # Calculates the total cost of the food plus tip
    record["total"] = round(record["subtotal"]*(1+(record["tip_perc"]/100)),digits=2);
    @printf("\nThe total cost after tipping is: \$%.2f",record["total"])
    print("\n")    

    # Determines if the bill would be split unevenly
    record["extra_person"] = round(record["total"]%0.05,digits=2)
    # If the bill is uneven then the user is pays for the extra 
    # cost ($0.04 at most)
    # Else, the bill is split evenly between the user and their # 4 friends
    if record["extra_person"] != 0.05
        record["per_person"] = (record["total"]-record["extra_person"])/5
        @printf("\nThe total cost for my friends is: \$%.2f",record["per_person"])
        @printf("\nThe total for me is: \$%.2f",record["per_person"]+record["extra_person"])
    else
        record["per_person"] = record["total"]/5
        @printf("\nThe total cost per person is: \$%.2f",record["per_person"])
    end

end
```
